# Blue Whale
A wrapper around the Bluesky public API.

```python
from bluewhale import BlueWhale

bw = BlueWhale()
```

## User Profile

```python
profile = bw.user_profile("forbes.com")

print(f"Display Name: {profile.displayName}")
print(f"Handle: {profile.handle}")
print(f"Followers: {profile.followersCount}")
```

## User Search

```python
results = bw.user_search("security researcher")

for user in results[:5]:
    print(f"@{user.handle} - {user.displayName}")
```

## User Feed

```python
feed = bw.user_feed("forbes.com", limit=5)

for item in feed:
    post = item.post
    print(f"@{post.author.handle}: {post.record.text[:80]}...")
```

## Trending Topics

```python
topics = bw.trending_topics(limit=10)

for topic in topics:
    print(f"#{topic.topic}")
```
